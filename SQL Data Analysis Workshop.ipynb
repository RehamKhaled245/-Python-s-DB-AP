```python
import mysql.connector
# Set up the database connection
db_connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="bike_store"
)
cursor = db_connection.cursor()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Function to import CSV data into a table
file_path="E:\depi"
def import_csv_to_table(file_path, table_name):
    # Read the file
    df = pd.read_csv(file_path)
  
    # Replace NaN values with None (NULL for MySQL)
    df = df.where(pd.notnull(df), None)
    
    # Prepare SQL placeholders and columns
    placeholders = ', '.join(['%s'] * len(df.columns))
    columns = ', '.join([f"`{col}`" for col in df.columns])
    sql = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"

    # Insert data into the table
    for row in df.itertuples(index=False):
        clean_row = tuple(None if pd.isna(value) else value for value in row)
        cursor.execute(sql, clean_row)
    db_connection.commit()
    print(f"Data from {file_path} inserted into {table_name}")

# Import data files dynamically
try:
    csv_files = [
        "brands.csv",
        "categories.csv",
        "products.csv",
        "stores.csv",
        "customers.csv",
        "staffs.csv",
        "orders.csv",
        "stocks.csv",
        "order_items.csv"
    ]
    
    for file in csv_files:
        table_name = file.split(".")[0]  # Extract table name from filename
        import_csv_to_table(file, table_name)
finally:
    cursor.close()
    db_connection.close()
    
```

### Task 1: Find the Store with the Highest Total Revenue
Retrieve the store name and total revenue (sum of list_price * quantity) across all orders.




```python
task1=pd.read_sql( """
    SELECT
    s.store_name AS StoreName, 
    SUM(oi.list_price * oi.quantity) AS TotalRevenue
FROM 
    stores s
JOIN 
    orders o ON s.store_id = o.store_id
JOIN 
    order_items oi ON o.order_id = oi.order_id
GROUP BY 
    s.store_name
ORDER BY 
    TotalRevenue DESC
LIMIT 1;""" ,db_connection)
task1
```

## Level 5: Additional Advanced SQL
(using Pandas read_sql)

### Task 1: Aggregate Sales by City and State
Write a query to calculate the total sales, average order value, and maximum order value for customers in each city and state.


```python
task1=pd.read_sql(
"""
select city,state,count(c.customer_id) as customer_count ,SUM( oi.list_price * oi.quantity)  AS total_sales ,AVG(list_price * quantity) AS average_order_value ,max(list_price * quantity) AS max_order_value from customers c 
inner join  orders o on c.customer_id=o.customer_id 
inner join order_items oi on oi.order_id=o.order_id
group by city
ORDER BY total_sales desc
""",db_connection)
task1
```

### Task 2: Find Top-Selling Products in 2018
Write a query to identify the products with the highest sales volume in the year 2018.




```python
task2 =pd.read_sql(
"""
select p.product_id,product_name,DATE_FORMAT(o.order_date, '%Y') AS order_year,sum( oi.list_price * oi.quantity / p.list_price)  as total_unit_sold from products p 
inner join order_items oi on oi.product_id = p.product_id
inner join orders o on o.order_id=oi.order_id
where YEAR(o.order_date) = 2018
group by product_name
order by total_unit_sold DESC
""",db_connection)
task2
```

### Task 3: Calculate Month-over-Month Sales Growth
Write a query to calculate the month-over-month sales growth for the past years.


```python
task3=pd.read_sql(
"""
select 
DATE_FORMAT(o.order_date, '%m') AS month ,DATE_FORMAT(o.order_date, '%Y') AS year,sum( oi.list_price * oi.quantity / p.list_price) as total_unit_sold ,SUM(oi.list_price * quantity) AS total_sales
 from orders o 
inner join order_items oi on o.order_id=oi.order_id
inner join products p on p.product_id=oi.product_id
GROUP BY 
    year, month
ORDER BY 
    year ASC, month ASC;
""",db_connection)
task4
```
